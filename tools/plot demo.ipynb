### Create the data

```python
x = np.linspace(0,10)
r = 10*random.random(np.size(x))
y = r + x**2 + 20.0
yerr = r
```

### Define a fit function

```python
def quad(x, k0, k1, k2):
    return k0 + x*k1 + x**2*k2
```

###Use the model fit command

Default: x, y, fitfunc, array of intial guess (all params need a guess) 
The default behavior is to fit with all the parameters as free parameters

```python
pt.plot_fit(x,y,quad,np.array([20,0,1]))
```

The function returns an array with the fit values and a second array with the uncertainties of each parameter (the sqrt of the diagonal elements of the covariance matrix, I think equivalent to the the 1 sigma uncertainty of the fit.

### Using the model hold option

It is nice to be able to quickly play with a model by just holding some parameters in a model. Define a boolean array where True is a parameter to hold fixed. I like to sometimes hold everything fixed to see how close my initial guesses are. Notice it tells you that it is not actually fitting anymore. I'm going to eventually change it so you can also pass 'all' to hold= to quickly check an intial guess.

```python
h = np.array([True, True, True])
pt.plot_fit(x,y,quad,np.array([0,0,1]),hold=h)
pt.plot_fit(x,y,quad,np.array([0,0,1]),hold=np.array([True,False,False]))
```

### Label and axis

```python
h = np.array([True, True, False])
pt.plot_fit(x,y,quad,np.array([0,0,1]),hold=h, labels=['myX','myY','model_fit.py demo'])
```
